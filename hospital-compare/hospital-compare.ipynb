# Finding the best hospital in a state

## Function to find the best hospital (lowest  30-day mortality)
Outcomes list:
1. Heart attack rate: col 11
2. Heart failure rate: col 17
3. Pneumonia rate: col 23

```r
best <- function(state, outcome_col) {
    col_list <- list(11,17,23)
    
    outcm <- read.csv("outcome-of-care-measures.csv",
                  na.strings='Not Available',
                  stringsAsFactors=FALSE,
                  colClasses = "character")
    if(!any(outcm[7] == state))
        return(message("Invalid state!"))
    
    if(!any(col_list == outcome_col))
        return(message("Invalid outcome!"))
        
    outcm[,outcome_col] <- as.numeric(outcm[,outcome_col])
    df_state <- outcm[outcm[7] == state, c(2,outcome_col)]
    result <- df_state[order(df_state[2], df_state[1]),]
    result[1,1]
}
```

## Best function test

```r
best(state='AL', outcome_col = 11)
```

# Rank hospital function

```r
rankhospital <- function(state, outcome_col, num = "best"){
    sorting <- if(num == 'best') {
        FALSE
    } else if (num == "worst") {
        TRUE
    } else if (is.numeric(num)){
        num
    } else {return("Invalid number / sorting method!")}
    
    col_list <- list(11,17,23)
    
    outcm <- read.csv("outcome-of-care-measures.csv",
                  na.strings='Not Available',
                  stringsAsFactors=FALSE,
                  colClasses = "character")
    
    if(!any(outcm[7] == state))
        return(message("Invalid state!"))
    if(!any(col_list == outcome_col))
        return(message("Invalid outcome!"))
    
    # No ranking specified
    if (sorting == TRUE || sorting == FALSE) {
        outcm[,outcome_col] <- as.numeric(outcm[,outcome_col])
        df_state <- outcm[outcm[7] == state, c(2,outcome_col)]
        result <- df_state[order(df_state[2], df_state[1], decreasing = sorting),]
        for(i in 1:nrow(result)){
            result$ranking[i] <- i
        }
        return (head(result))
    }
    
    # Ranking specified
    if (is.numeric(sorting)){
        outcm[,outcome_col] <- as.numeric(outcm[,outcome_col])
        df_state <- outcm[outcm[7] == state, c(2,outcome_col)]
        result <- df_state[order(df_state[2], df_state[1], decreasing = FALSE),]
        for(i in 1:nrow(result)){
            result$ranking[i] <- i
        }
        return (result[sorting, ])
    }
}
```

```r
rankhospital('AL', 11, 20)
```

# Ranking hospitals in all states

```r
rankall <- function(outcome_col, num = "best") {
    sorting <- if(num == 'best') {
        FALSE
    } else if (num == "worst") {
        TRUE
    } else if (is.numeric(num)){
        num
    } else {return("Invalid number / sorting method!")}
    
    col_list <- list(11,17,23)
    
    outcm <- read.csv("outcome-of-care-measures.csv",
                  na.strings='Not Available',
                  stringsAsFactors=FALSE,
                  colClasses = "character")
    
    if(!any(col_list == outcome_col))
        return(message("Invalid outcome!"))
    # No ranking specified
    if (sorting == TRUE || sorting == FALSE) {
        sorted_df <- outcm[order(outcm[7], outcm[outcome_col], outcm[2], decreasing = sorting), c(7,outcome_col, 2)]
        groupby_state_df <- split(sorted_df[2], sorted_df[1])
        return (lapply(groupby_state_df, function(x) return (x[1:6,])))
    }
                   
    # Ranking specified
    if (is.numeric(sorting)){
        sorted_df <- outcm[order(outcm[7], outcm[outcome_col], decreasing = FALSE), c(7,outcome_col)]
        groupby_state_df <- split(sorted_df[2], sorted_df[1])
        return (unlist(lapply(groupby_state_df, function(x) return (x[sorting,]))))
    }
}
```
